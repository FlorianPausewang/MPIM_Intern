```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

d=np.loadtxt("./GPM00078897-drvd-1980-1990.txt")
```

```python
#np.set_printoptions(threshold=)
ds=d[:,[0,-7]]
print(ds)
```

```python


def sort(data,ds,row):
    
    if row[1]==-99999:
        #print("Nan")
        return data
    
    for plev in data:
        #print('here',plev,plev[0],0.01*row[0])

        if plev[0]==0.01*row[0]:
            #print("insert")
            new=np.array([0.001*row[1]])
            plev[1]=np.append(plev[1],new,axis=0)
            return data
            
        elif plev[0]==data[-1,0]:
            #print("new")
            new=np.array([[0.01*row[0],np.array([0.001*row[1]])]])
            data=np.append(data,new,axis=0)
            return data
        

```

```python
#Create new array

row=ds[0]
data=np.array([[0.01*row[0],np.array([0.001*row[1]])]])
print(data)

#Sort
for row in ds:
        
        
        
        data=sort(data,ds,row)
        #if len(data[:,])>10:
        #    break

            
print(data)           
```

```python
sorted_indices = np.argsort(-data[:, 0])

# Use the indices to reorder the array
data = data[sorted_indices]

print(data)
```

```python
fig, ax = plt.subplots()


ax.plot(data[:,0],[len(arr) for arr in data[:,1]])
#ax.set_xlim((395,405))
#ax.set_xlim((975,1025))

ax.set_yscale('log')
#ax.set_title('Zonal mean Temperature in different heights p')
ax.grid()
ax.legend()#title = '')
plt.savefig('NumberofValues.png',dpi=400)
plt.show()
```

```python
p = data[:,0]
means=np.array([np.mean(arr) for arr in data[:,1]])
stds=np.array([np.std(arr) for arr in data[:,1]])


mask = np.array([len(arr) >= 50 for arr in data[:,1]])

print(p,means)

p_bins=np.flip(np.arange(0,1050,25))
print(p_bins)
rh_bins=[[] for p in p_bins]


for plev in data:
    i=0
    while ( p_bins[i]-(25/2) )>plev[0]:
        i=i+1
        
    rh_bins[i]=np.append(rh_bins[i],means[i])  
    
print(rh_bins)
```

```python

fig, ax = plt.subplots()


ax.plot(p_bins,[len(arr) for arr in rh_bins])
#ax.set_xlim((395,405))
#ax.set_xlim((975,1025))

ax.set_yscale('log')
#ax.set_title('Zonal mean Temperature in different heights p')
ax.grid()
ax.legend()#title = '')
plt.savefig('NumberofValues.png',dpi=400)
plt.show()

```
